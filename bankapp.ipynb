```python
balance= 0
data={2323: 'Segun'}
def openup():
    print('''Hello, welcome to FinTech.
    Click. 1 to create an account.
    Click. 2 to start a transaction
    Click, 3 to exit''')
    create= int(input('Enter a number\n'))
    if create == 1:
        username= input('Enter your name\n')
        password= int(input('Enter your password\n'))
    elif create == 2:
        transact()
    elif create == 3:
        print('Goodbye. See you later')
        sys.exit()
        




def transact():
    password= int(input('Enter Your Numeric Password\n'))
    if password in data:
        print('''Welcome!
        Press 1, To Check Balance\n
        Press 2, Make a deposit\n
        Press 3, To Withdraw\n
        Press 4, To Transfer\n
        Press 5, To exit \n''')
        openaccount= int(input('Enter a number\n'))
        if openaccount == 1:
            checkbalance()
        elif openaccount == 2:
            deposit()
        elif openaccount == 3:
            withdraw()
        elif openaccount == 4:
            transfer()
        elif openaccount == 5:
            print('Good bye, see you next time')
    elif password not in data:
        print('Invalid password\n', transact())

def checkbalance():
    print('Your account balance is:\n', balance)

def deposit():
    try:
        depo= float(input('Type in the amount to deposit\n'))
        global balance
        balance += depo
        print('\n You deposited:\n', depo)
    except ValueError:
        print('Oops! Something went wrong. Try again\n', deposit())

def withdraw():
    try:
        amount= float(input('Enter amount to withdraw\n'))
        global balance
        if balance >= amount: 
            balance -= amount
            print('You withdrew:\n', amount, transact())
        else:
            print('Insufficient balance, make a deposit.\n')
    except ValueError:
        print('Enter numbers\n', deposit())

def transfer():
    try:
        trans= float(input('Type in amount to transfer:\n'))
        global balance
        if balance >= trans:
            balance -= trans
            print('You transferred:\n', trans, transact())
        else:
            print('Inssuficient balance, make a deposit.\n', deposit())
    except ValueError:
        print('Oops! Type in numbers please', transfer())

        
    
            
            
            
    
        
    
        
            
```
